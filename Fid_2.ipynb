```python
#Here, we introduce the necessary libraries that are essential for our purpose.

import pennylane as qml
from pennylane import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Here, 'wires' represent the total number of qubits, 
# and 'default.mixed' is the device necessary for simulating noisy circuits.
dev=qml.device('default.mixed',wires=4)
```

```python
# Proj is used to apply a projector on the final states to take measurements.
# Although PennyLane supports qml.unitary only for unitary matrix implementation but one can use ti apoply a projector as well it works fine.
proj = np.array([[1, 0], [0, 0]])

#Here wen defined quantum node esstinal to run pennylane circuit 
@qml.qnode(dev)


def circuit2(params,p):
# this is how one can obtain phi_plus 
    qml.Hadamard(0)
    qml.CNOT(wires=[0,1])
    
    
# Implementation of U
    qml.U3(params[0],params[1],params[2],wires=1)
    qml.U3(params[3],params[4],params[5],wires=2)
    qml.CNOT(wires=[2,1])
    qml.RZ(params[6],wires=1)
    qml.RY(params[7],wires=2)
    qml.CNOT(wires=[1,2])
    qml.RY(params[8],wires=2)
    qml.CNOT(wires=[2,1])
    qml.U3(params[9],params[10],params[11],wires=1)
    qml.U3(params[12],params[13],params[14],wires=2)

    qml.RZ(params[15],wires=[3])
    qml.CNOT(wires=[2,3])
    qml.RZ(params[16],wires=[3])
    qml.CNOT(wires=[1,3])
    qml.RZ(params[17],wires=[3])
    qml.CNOT(wires=[2,3])
    qml.RZ(params[18],wires=[3])
    qml.CNOT(wires=[1,3])


    qml.U3(params[19],params[20],params[21],wires=1)
    qml.U3(params[22],params[23],params[24],wires=2)
    qml.CNOT(wires=[2,1])
    qml.RZ(params[25],wires=1)
    qml.RY(params[26],wires=2)
    qml.CNOT(wires=[1,2])
    qml.RY(params[27],wires=2)
    qml.CNOT(wires=[2,1])
    qml.U3(params[28],params[29],params[30],wires=1)
    qml.U3(params[31],params[32],params[33],wires=2)

    qml.RY(params[34],wires=[3])
    qml.CNOT(wires=[2,3])
    qml.RY(params[35],wires=[3])
    qml.CNOT(wires=[1,3])
    qml.RY(params[36],wires=[3])
    qml.CNOT(wires=[2,3])
    qml.RY(params[37],wires=[3])
    qml.CNOT(wires=[1,3])


    qml.U3(params[38],params[39],params[40],wires=1)
    qml.U3(params[41],params[42],params[43],wires=2)
    qml.CNOT(wires=[2,1])
    qml.RZ(params[44],wires=1)
    qml.RY(params[45],wires=2)
    qml.CNOT(wires=[1,2])
    qml.RY(params[46],wires=2)
    qml.CNOT(wires=[2,1])
    qml.U3(params[47],params[48],params[49],wires=1)
    qml.U3(params[50],params[51],params[52],wires=2)

    qml.RZ(params[53],wires=[3])
    qml.CNOT(wires=[2,3])
    qml.RZ(params[54],wires=[3])
    qml.CNOT(wires=[1,3])
    qml.RZ(params[55],wires=[3])
    qml.CNOT(wires=[2,3])
    qml.RZ(params[56],wires=[3])
    qml.CNOT(wires=[1,3])


    qml.U3(params[57],params[58],params[59],wires=1)
    qml.U3(params[60],params[61],params[62],wires=2)
    qml.CNOT(wires=[2,1])
    qml.RZ(params[63],wires=1)
    qml.RY(params[64],wires=2)
    qml.CNOT(wires=[1,2])
    qml.RY(params[65],wires=2)
    qml.CNOT(wires=[2,1])
    qml.U3(params[66],params[67],params[68],wires=1)
    qml.U3(params[69],params[70],params[71],wires=2)
    
    
#Implimentation of Noise 
    qml.DepolarizingChannel(1-p,wires=1)
    qml.DepolarizingChannel(1-p,wires=2)
    qml.DepolarizingChannel(1-p,wires=3)
    
    
#Implementation of U_dagger 
    qml.adjoint(qml.U3(params[69],params[70],params[71],wires=2))
    qml.adjoint(qml.U3(params[66],params[67],params[68],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.RY(params[65],wires=2))
    qml.CNOT(wires=[1,2])
    qml.adjoint(qml.RY(params[64],wires=2))
    qml.adjoint(qml.RZ(params[63],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.U3(params[60],params[61],params[62],wires=2))
    qml.adjoint(qml.U3(params[57],params[58],params[59],wires=1))

    qml.CNOT(wires=[1,3])
    qml.adjoint(qml.RZ(params[56],wires=[3]))
    qml.CNOT(wires=[2,3])
    qml.adjoint(qml.RZ(params[55],wires=[3]))
    qml.CNOT(wires=[1,3])
    qml.adjoint(qml.RZ(params[54],wires=[3]))
    qml.CNOT(wires=[2,3])
    qml.adjoint(qml.RZ(params[53],wires=[3]))

    qml.adjoint(qml.U3(params[50],params[51],params[52],wires=2))
    qml.adjoint(qml.U3(params[47],params[48],params[49],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.RY(params[46],wires=2))
    qml.CNOT(wires=[1,2])
    qml.adjoint(qml.RY(params[45],wires=2))
    qml.adjoint(qml.RZ(params[44],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.U3(params[41],params[42],params[43],wires=2))
    qml.adjoint(qml.U3(params[38],params[39],params[40],wires=1))

    qml.CNOT(wires=[1,3])
    qml.adjoint(qml.RY(params[37],wires=[3]))
    qml.CNOT(wires=[2,3])
    qml.adjoint(qml.RY(params[36],wires=[3]))
    qml.CNOT(wires=[1,3])
    qml.adjoint(qml.RY(params[35],wires=[3]))
    qml.CNOT(wires=[2,3])
    qml.adjoint(qml.RY(params[34],wires=[3]))

    qml.adjoint(qml.U3(params[31],params[32],params[33],wires=2))
    qml.adjoint(qml.U3(params[28],params[29],params[30],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.RY(params[27],wires=2))
    qml.CNOT(wires=[1,2])
    qml.adjoint(qml.RY(params[26],wires=2))
    qml.adjoint(qml.RZ(params[25],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.U3(params[22],params[23],params[24],wires=2))
    qml.adjoint(qml.U3(params[19],params[20],params[21],wires=1))

    qml.CNOT(wires=[1,3])
    qml.adjoint(qml.RZ(params[18],wires=[3]))
    qml.CNOT(wires=[2,3])
    qml.adjoint(qml.RZ(params[17],wires=[3]))
    qml.CNOT(wires=[1,3])
    qml.adjoint(qml.RZ(params[16],wires=[3]))
    qml.CNOT(wires=[2,3])
    qml.adjoint(qml.RZ(params[15],wires=[3]))

    qml.adjoint(qml.U3(params[12],params[13],params[14],wires=2))
    qml.adjoint(qml.U3(params[9],params[10],params[11],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.RY(params[8],wires=2))
    qml.CNOT(wires=[1,2])
    qml.adjoint(qml.RY(params[7],wires=2))
    qml.adjoint(qml.RZ(params[6],wires=1))
    qml.CNOT(wires=[2,1])
    qml.adjoint(qml.U3(params[3],params[4],params[5],wires=2))
    qml.adjoint(qml.U3(params[0],params[1],params[2],wires=1))
    
    
    
#Post selection 
    qml.QubitUnitary(proj,wires=2)
    qml.QubitUnitary(proj,wires=3)

    
    
# Return the density matrix for the first two qubits.
# This will perform a partial trace over the third and forth system, effectively tracing out the third and forth qubit.
    return qml.density_matrix(wires=[0, 1])
```

```python
#This section of the code is solely for visualizing the circuit.
p=0.6
params=np.random.uniform(-np.pi,np.pi,72)
fig,ax=qml.draw_mpl(circuit2)(params,p)
plt.savefig("Fid_2ancilla")
```

```python
# Cost Function Definition: Typically, optimizers are minimizers, so we subtract the cost function from one to reverse the process, effectively making the optimizer a maximizer.
def fidelity(state, target):
    target_dm = np.outer(target, np.conj(target))
    fidd = np.abs(np.dot(target.conj().T, np.dot(state, target)))
    fid=fidd/np.trace(state)
    return np.real(fid)
target_state = np.array([1, 0, 0, 1]) / np.sqrt(2)
def cost2(params,p):
    return 1-fidelity(circuit2(params,p), target_state)
```

```python
# Define the optimizer 
optimizer = qml.GradientDescentOptimizer(stepsize=0.1)
all_success_probabilities2 = []
all_fidelities2 = []
p_values = [0.25,0.3,0.35,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]
num_seeds = 5

# Optimization loop
for p in p_values:
    fid_init2 = 0
    p_sucss_init2 = 0
    for seed in range(num_seeds):

        params=np.random.uniform(-np.pi,np.pi,72,requires_grad=True)
        
        for i in range(150):
            params = optimizer.step(lambda u: cost2(u, p), params)
        
        circuit_output2 = circuit2(params, p)
        current_fidelity2 = fidelity(circuit_output2, target_state)
        print(current_fidelity2)
        current_success_prob2 = np.trace(circuit_output2).real  # Assuming circuit2 returns a density matrix
        
        if current_fidelity2 > fid_init2:
            fid_init2 = current_fidelity2
            p_sucss_init2 = current_success_prob2
    
    final_fid2 = fid_init2
    final_p2 = p_sucss_init2
    all_fidelities2.append(final_fid2)
    all_success_probabilities2.append(final_p2)
print(all_fidelities2)
print(all_success_probabilities2)
```
